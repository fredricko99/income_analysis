```python
import numpy as np
import pandas as pd
from numpy import array
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
```

**Income Data**

```python
population_df = pd.read_csv('https://raw.githubusercontent.com/Explore-AI/Public-Data/master/AnalyseProject/world_population.csv', index_col='Country Code')
meta_df = pd.read_csv('https://raw.githubusercontent.com/Explore-AI/Public-Data/master/AnalyseProject/metadata.csv', index_col='Country Code')
```

```python
def get_total_pop_by_income(income_group_name='Low income'):
    
    '''
    parameters: 
    income group name: name of the income group to get total income
    '''
    
    _df = population_df.copy().join(meta_df)
    if income_group_name not in _df['Income Group'].unique():
        raise ValueError('Specified income group: "{}", not found in DataFrame'.format(income_group_name))
    _df = _df[_df['Income Group'] == income_group_name].drop(['Region', 'Income Group', 'Special Notes'], axis=1)
    _df = _df.sum()
    _df = _df.astype(np.int64)
    X = list(map(np.int64, population_df.columns))
    Y = np.array(_df)
    return np.vstack([X,Y]).T
```

```python
data = get_total_pop_by_income('High income')
data
```

```python
def sklearn_kfold_split(data,K):
    '''
    takes as input a 2-d numpy array and an integer `K` corresponding to the number of splits. This function will then return a list of tuples of length `K`. Each tuple in this list consists of a `train_indices` list and a `test_indices` list containing the training/testing data point indices for that particular $K^{th}$ split. 
    parameters:
    data: the 2Darray containing data 
    K: the number of splits to be done
    '''
    kfold = KFold(n_splits=K)
    data_i = [(train_list,test_list) for train_list,test_list in kfold.split(data)]
    return list(data_i)
```

```python
data = get_total_pop_by_income('High income');
sklearn_kfold_split(data,4)
```

```python
def best_k_model(data,data_indices):
    m_list = []
    m_mse = []
    for train_indices,test_indices in data_indices:
        
        # init model
        model = RandomForestRegressor(random_state=42)
        X_train, y_train = data[train_indices,0],data[train_indices,1]
        X_test, y_test = data[test_indices,0],data[test_indices,1]
        
        # train model
        model.fit(X_train[:,np.newaxis], y_train) 
        m_list.append(model)
        
        # test model
        mse = mean_squared_error(model.predict(X_test[:,np.newaxis]),y_test)
        m_mse.append(mse)
    
    # get index of highest mse
    best_model_ = np.argmax(m_mse)
    
    return m_list[best_model_]
```

```python
data = get_total_pop_by_income('High income')
data_indices = sklearn_kfold_split(data,5)

best_model = best_k_model(data,data_indices)
best_model.predict([[1960]])
```
